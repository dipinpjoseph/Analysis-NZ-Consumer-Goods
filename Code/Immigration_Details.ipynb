```r
#Loading Libraries
library(tidyverse)
library(magrittr) # better handling of pipes
library(purrr) # to work with lists and map functions
library(glue) # to paste strings
library(stringr) # to hand strings
library(rvest) # rvest makes scraping easier
library(polite) # polite is the "polite" version of rvest
```

```r

url <- "https://en.wikipedia.org/wiki/Immigration_to_New_Zealand"  # this is the base url from where the scraping starts
Immigration <- read_html(url) %>%
html_table(fill=TRUE)
```

```r
#Immigration_details

df_Immigration_details <- data.frame(Immigration[1])
Immigration_details <- df_Immigration_details %>%
filter(Country !="Country") %>%
select(Country,Gross.arrivals,Gross.arrivals.1,Gross.arrivals.2,Gross.arrivals.3,Gross.arrivals.4) %>%
rename(Gross.arrivals_2013=Gross.arrivals,Gross.arrivals_2014=Gross.arrivals.1,
       Gross.arrivals_2015=Gross.arrivals.2,Gross.arrivals_2016=Gross.arrivals.3,
       Gross.arrivals_2017=Gross.arrivals.4)

Immigration_details
```
