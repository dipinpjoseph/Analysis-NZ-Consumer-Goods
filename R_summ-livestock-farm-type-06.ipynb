```r
library(tidyverse)
library(readxl)
```

```r
#setting the file name
summ_livestock_file <- "summ-livestock-farm-type-06.xls"
```

```r
summ_livestock_file %>%
    read_excel(sheet = "Livestock by farm type", range = 'A7:L36', col_names = c("Farm_Type", "Missing", "Dairy_Cattle", "Beef_Cattle", "Sheep", "Deer", "Pigs", "Goats", "Horses", "Ostriches_Emus", "Alpacas_Ilamas", "All_Other_Livestock")) -> table_1_untidy

table_1_untidy
```

```r
table_1_untidy1 <- table_1_untidy[ , -2]
table_1_untidy1
```

```r
table_1_untidy1 %>%
    replace(table_1_untidy1 == "-", 0)%>%
    replace(table_1_untidy1 == "C", 0) -> table_1_tidy2

table_1_tidy2
```
